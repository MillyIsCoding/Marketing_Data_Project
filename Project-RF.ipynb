**Categorical Variables**

```python
## Finding Unique Values of Categorical Columns + Dealing with the data accodingly
year = df["Year_Birth"].unique()
year.sort()
print(year)
```

```python
mStatus = df['Marital_Status'].unique()
mStatus.sort()
print(mStatus)
totals = df["Marital_Status"].value_counts()
print(totals)
```

```python
# Alone values can be changed to Single. Absurd and YOLO arent real marital status' and can be removed due their small contribution to the dataset.
df["Marital_Status"] = df["Marital_Status"].replace('Alone','Single')
df = df[~df["Marital_Status"].isin(["Absurd","YOLO"])]
print(df["Marital_Status"].value_counts())
```

```python
##Unique Variables for Education Column
edu = df['Education'].unique()
print(edu)
print(df['Education'].value_counts())
```

```python
#Label Encoding Education Column to preserve rank of education level
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education']).astype(int)
```

```python
##One-Hot Encoding Marital_Status
df = pd.get_dummies(df, columns=['Marital_Status'])
```

```python
#Drop Columns without neccesary information
df.drop(columns=["ID"],inplace=True)
df
```

Using this copy of the dataset to create a baseline model on usable information

**Due to its formatting we cannot use the Dt_customer column as part of our baseline model, we will reinstitute the information in the column during the feature engineering phase**

```python
df_baseline = df_baseline.drop(columns=['Dt_Customer'])
```

```python
X = df_baseline.drop(columns=['Response'])
y = df_baseline['Response']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

classifier = DecisionTreeClassifier(random_state=0)

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
ConfusionMatrixDisplay(conf_matrix).plot();
```

```python
BaseMetrics = {
    'accuracy':accuracy_score(y_test, y_pred),
    'precision':precision_score(y_test, y_pred),
    'recall':recall_score(y_test, y_pred),
    'F1-Score':f1_score(y_test, y_pred)
}
print('Baseline Metrics')
for key,value in BaseMetrics.items():
    print(f'{key}:{value}')
```

**Because the goal of this model is to reach as many potential customers as possible Our Focus for improving the model will be Recall and F1**

```python
BaseMetrics = {
    'recall':recall_score(y_test, y_pred),
    'F1-Score':f1_score(y_test, y_pred)
}
```

```python
df2 = df.copy()
df2
```

```python
##Addressing Outliers using IQR
Q1 = df2['Income'].quantile(0.25)
Q3 = df2['Income'].quantile(0.75)
IQR = Q3 - Q1

#Identify the outliers in the Income column
outliers = df2[(df2['Income'] < (Q1 - 1.5 * IQR)) | (df2['Income'] > (Q3 + 1.5 * IQR))]

#Print the number of outliers
print("Outliers in the Income column:", len(outliers))

df2 = df2[~((df2['Income'] < (Q1 - 1.5 * IQR )) | (df2['Income'] > (Q3 + 1.5 * IQR )))]

Q1 = df2['Year_Birth'].quantile(0.25)
Q3 = df2['Year_Birth'].quantile(0.75)
IQR = Q3 - Q1

#Identify the outliers in the Year_Birth column
outliers = df2[(df2['Year_Birth'] < (Q1 - 1.5 * IQR)) | (df2['Year_Birth'] > (Q3 + 1.5 * IQR))]

#Print the number of outliers
print("Outliers in the Year_Birth column:", len(outliers))

df2 = df2[~((df2['Year_Birth'] < (Q1 - 1.5 * IQR )) | (df2['Year_Birth'] > (Q3 + 1.5 * IQR )))]
```

```python
#Feature Engineering
# Create a new column named 'Age'.
df2['Age'] = 2025 - df2['Year_Birth']

# Create a new column for all the accepted campaigns.
df2['Accepted_Campaigns'] = df2['AcceptedCmp1'] + df2['AcceptedCmp2'] + df2['AcceptedCmp3'] + df2['AcceptedCmp4'] + df2['AcceptedCmp5']

# Create a new column for total items purchased and removing columns to simplify data.
df2['Total_Items'] = df2['MntWines'] + df2['MntFruits'] + df2['MntMeatProducts'] + df2['MntFishProducts'] + df2['MntSweetProducts'] + df2['MntGoldProds']


# Create a new column for all the purchases.
df2['Total_Purchases'] = df2['NumDealsPurchases'] + df2['NumWebPurchases'] + df2['NumCatalogPurchases'] + df2['NumStorePurchases']

##Not useful data column
df2=df2.drop(columns=['Dt_Customer','Z_CostContact','Z_Revenue','Recency','Complain','Year_Birth']) 
```

**Retune Model**

```python
X = df2.drop(columns=['Response'])
y = df2['Response']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

classifier = DecisionTreeClassifier(random_state=0)

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

**Hyper Parameter Tuning of Classification Model**

```python
param_grid = {
    'max_depth': [3, 5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy'],
    'max_features': [None, 'sqrt', 'log2']
}

scoring = {
    'recall': 'recall'
}

grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=0),
    param_grid=param_grid,
    scoring=scoring,
    refit='recall', 
    cv=5,
)

grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Recall Score:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_

# Predict on test data
y_pred = best_model.predict(X_test)

# Evaluate the model
TunedMetrics = {
'recall':recall_score(y_test, y_pred_rf),
'F1-Score':f1_score(y_test, y_pred)
}
```
