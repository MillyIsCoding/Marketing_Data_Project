```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("marketing_campaign.csv",sep=";")
df
```

```python
df["Year_Birth"].hist()
```

The distribution of the birth years is normal but there is a clear 

```python
for status, group in df.groupby("Marital_Status"):
    group["Income"].hist(alpha=0.5, label=status, bins=30)

plt.legend()
plt.xlabel("Income")
plt.title("Income Distribution by Marital Status")
plt.show()
```

The distributions of income when based on marital status is normal throughout the categories but there are outliers present

```python
for status, group in df.groupby("Education"):
    group["Income"].hist(alpha=0.5, label=status, bins=30)

plt.legend()
plt.xlabel("Income")
plt.title("Income Distribution by Education")
plt.show()
```

-Majority of customer have at least graduated college

-Distributions are fairly normal throughout the different levels but there are some clear outliers in a few of the categories
